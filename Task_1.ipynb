```python
scaler = StandardScaler()
def get_features_and_target_arrays(df,numerical_cols,cat_cols,scaler):
  x_numeric_scaled = scaler.fit_transform(df[numerical_cols])
  x_categorical = df[cat_cols].to_numpy()
  x = np.hstack(( x_categorical, x_numeric_scaled))
  y = df["target"]

  return x,y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of each numerical feature
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Boxplot for each numerical feature
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
corr_matrix = df[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Count plot for the target variable
plt.figure(figsize=(8, 4))
sns.countplot(x='target', data=df)
plt.title('Target Variable Distribution')
plt.show()
```
